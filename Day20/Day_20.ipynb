# Part 1

```python
from collections import deque
from typing import Literal

file_name = "input.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

G = [[c for c in line] for line in data]
R =len(G)
C = len(G[0])

for r in range(R):
    for c in range(C):
        if G[r][c] == "S":
            start_r, start_c = r,c
        if G[r][c] == "E":
            end_r, end_c = r,c
    
def bfs(G, start_r, start_c, end_r, end_c, max_d = None):
    
    visited = set()
    queue = deque()
    queue.append((0, start_r, start_c))

    while queue:
        d,r,c= queue.popleft()
        
        if max_d is not None:
            if d > max_d:
                return -1

        if (r,c) in visited:
            continue
        visited.add((r,c))
        
        if r == end_r and c == end_c:
            return(d)

        # find neighbors
        for dr,dc in [(-1,0),(1,0),(0,-1),(0,1)]:
            rr = r+dr
            cc = c+dc
            if 0<=rr<R and 0<=cc<C and G[rr][cc] != "#" and (rr,cc) not in visited:
                queue.append((d+1,rr, cc))
    return -1

reference_dist  = bfs(G, start_r, start_c, end_r, end_c, max_d = None)
```

```python
shorter_paths = []
for r in range(R):
    for c in range(C):
        if r>0 and c>0 and r<R-1 and c<C-1:
            if G[r][c] == "#":
                G[r][c] = "."
                new_d = bfs(G, start_r, start_c, end_r, end_c, max_d = reference_dist-100)
                if new_d > 0:
                    shorter_paths.append(new_d)
                G[r][c] = "#"

print(len(shorter_paths))
```
