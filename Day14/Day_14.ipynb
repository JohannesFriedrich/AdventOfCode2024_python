# Part 1

```python
import re
file_name = "input.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

robots = []
for robot in data:
    match = re.match(r"p=(?P<x>\d+),(?P<y>\d+) v=(?P<vx>-?\d+),(?P<vy>-?\d+)", robot)
    t = {k: int(v) for k,v in match.groupdict().items()}
    robots.append(t)

R = 103
C = 101

def next_step(robot):
    robot["x"] = (robot["x"] + robot["vx"])%C
    robot["y"] = (robot["y"] + robot["vy"])%R
    return robot

# calc positions after 100 s
for _ in range(100):
    for i,robot in enumerate(robots):
        robot = next_step(robot)
    robots[i] = robot

# split grid    
quadrant = {1:0, 2:0, 3:0, 4:0}
for robot in robots:
    if 0 <= robot["x"] < C//2 and 0 <= robot["y"] < R//2:
        quadrant[1] += 1
    if 0 <= robot["x"] < C//2 and R//2+1 <= robot["y"] < R:
        quadrant[4] += 1
    if C//2+1 <= robot["x"] < C and 0 <= robot["y"] < R//2:
        quadrant[2] += 1
    if C//2+1 <= robot["x"] < C and R//2+1 <= robot["y"] < R:
        quadrant[3] += 1
    
res = 1
for q in quadrant.values():
    res *= q
print(res)
```

```python
from copy import deepcopy
import re
from collections import deque
file_name = "input.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

robots = []
for robot in data:
    match = re.match(r"p=(?P<x>\d+),(?P<y>\d+) v=(?P<vx>-?\d+),(?P<vy>-?\d+)", robot)
    t = {k: int(v) for k,v in match.groupdict().items()}
    robots.append(t)

R = 103
C = 101
G = [["."]*C for r in range(R)]

def next_step(robot):
    robot["x"] = (robot["x"] + robot["vx"])%C
    robot["y"] = (robot["y"] + robot["vy"])%R
    return robot


def split_grid(robots):
    quadrant = [[] for i in range(5)]
    for robot in robots:
        if 0 <= robot["x"] < C//2 and 0 <= robot["y"] < R//2:
            quadrant[1].append((robot["x"], robot["y"]))
        if C//2+1 <= robot["x"] < C and 0 <= robot["y"] < R//2:
            quadrant[2].append((robot["x"], robot["y"]))
        if C//2+1 <= robot["x"] < C and R//2+1 <= robot["y"] < R:
            quadrant[3].append((robot["x"], robot["y"]))
        if 0 <= robot["x"] < C//2 and R//2+1 <= robot["y"] < R:
            quadrant[4].append((robot["x"], robot["y"]))
    return quadrant


def bfs(start, visited):

    queue = deque()
    queue.append((start[0], start[1],0))

    while queue:
        r, c, d = queue.popleft()

        if (r,c) in visited:
            continue

        visited.add((r,c))  

        for dr,dc in [(1,0),(-1,0),(0,1),(0,-1)]:
            rr = r + dr
            cc = c + dc
            if 0 <= rr < R and 0 <= cc < C and H[rr][cc] == "#" and (rr,cc) not in visited:
                queue.append((rr,cc,d+1))

    return visited

seconds = 0
found = False
while True:
    if found:
        break
    seconds += 1
    H = deepcopy(G)
    coords = set()
    for i,robot in enumerate(robots):
        robots[i] = next_step(robot)
        x,y  = robots[i]["y"], robots[i]["x"]
        H[x][y] = "#"
        coords.add((x,y))
        
    while coords:
        visited = set()
        start = [v for i,v in enumerate(coords) if i == 0][0]
        visited = bfs(start, visited)
        if len(visited) > 200:
            print(f"Found 200 connected robots after {seconds} s")
            found = True
        coords = coords - visited

# 6355
```
