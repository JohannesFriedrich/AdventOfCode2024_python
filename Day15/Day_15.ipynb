# Part 1

```python
import re

file_name = "input.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n\n")

G, moves = data
G = [[c for c in line] for line in G.split("\n")]
moves = re.sub("\n", "", moves)
R =len(G)
C = len(G[0])

DIRS = {
    ">": (0,1),
    "^": (-1,0),
    "<": (0,-1),
    "v": (1,0)}

walls = []
boxes = []
for r in range(R):
    for c in range(C):
        if G[r][c] == "@":
            robot = (r,c)
        if G[r][c] == "#":
            walls.append((r,c))
        if G[r][c] == "O":
            boxes.append((r,c))

def execute_box_move(move,G,robot,rr,cc):
    if move == ">":  #done
        row = G[rr][cc:]
        if "." in row:
            next_dot = row.index(".")
            next_wall = row.index("#")
            if next_wall > next_dot: # there is a free space in the row
                G[rr][cc+row.index(".")] = "O"
                G[rr][cc] = "@"
                G[rr][cc-1] = "."
                robot = (rr,cc)
    if move == "<": # done
        row = G[rr][:cc]
        if "." in row:
            next_dot = [i for i, x in enumerate(row) if x == "."][-1]
            next_wall = [i for i, x in enumerate(row) if x == "#"][-1]
            if next_wall < next_dot:
                G[rr][next_dot] = "O"
                G[rr][cc] = "@"
                G[rr][cc+1] = "."
                robot = (rr,cc)
    if move == "v": #done
        col = [l[cc] for l in G[rr:]]
        if "." in col: # there is a free space in the row
            next_dot = col.index(".")
            next_wall = col.index("#")
            if next_wall > next_dot:
                G[rr+next_dot][cc] = "O"
                G[rr][cc] = "@"
                G[rr-1][cc] = "."
                robot = (rr,cc)
    if move == "^":
        col = [l[cc] for l in G[:rr]]
        if "." in col: # there is a free space in the row
            next_dot = [i for i, x in enumerate(col) if x == "."][-1]
            next_wall = [i for i, x in enumerate(col) if x == "#"][-1]
            if next_wall < next_dot:
                G[next_dot][cc] = "O"
                G[rr][cc] = "@"
                G[rr+1][cc] = "."
                robot = (rr,cc)
       
    return G, robot


for move in moves:
    rr, cc = robot[0] + DIRS[move][0], robot[1] + DIRS[move][1]
    if (rr,cc) not in walls and G[rr][cc] == ".":
        G[robot[0]][robot[1]] = "."
        robot = (rr,cc)
        G[rr][cc] = "@"
    if G[rr][cc] == "O":
        G, robot = execute_box_move(move,G,robot, rr,cc)

res = 0
for r in range(R):
    for c in range(C):
        if G[r][c] == "O":
            res += r*100 + c

print(res)   
```
