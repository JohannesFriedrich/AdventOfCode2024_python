```python
def find_all_paths(maze, start, end):
    # Definiere die möglichen Bewegungsrichtungen (nach oben, unten, links, rechts)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    def is_valid_move(x, y):
        # Prüfe, ob der nächste Schritt innerhalb des Labyrinths und nicht auf einer Wand ist
        return 0 <= x < len(maze) and 0 <= y < len(maze[0]) and maze[x][y] != '#'

    def dfs(x, y, path):
        # Wenn wir das Ziel erreicht haben, speichern wir den aktuellen Pfad
        if (x, y) == end:
            paths.append(list(path))
            return
        
        # Markiere den aktuellen Punkt als besucht, um Schleifen zu vermeiden
        maze[x][y] = '#'
        
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if is_valid_move(nx, ny):
                path.append((nx, ny))  # Gehe zum nächsten Punkt
                dfs(nx, ny, path)      # Rekursive DFS-Funktion aufrufen
                path.pop()             # Backtrack, d.h. gehe zurück und versuche den nächsten Weg

        # Markiere den aktuellen Punkt wieder als unbesucht für andere Pfade
        maze[x][y] = '.'

    # Initialisiere die Liste der gefundenen Wege
    paths = []
    # Starte DFS von Startpunkt S
    sx, sy = start
    maze[sx][sy] = '#'
    dfs(sx, sy, [start])

    return paths


# Definiere das Maze (das Labyrinth)
maze = [
    list("###############"),
    list("#.......#....E#"),
    list("#.#.###.#.###.#"),
    list("#.....#.#...#.#"),
    list("#.###.#####.#.#"),
    list("#.#.#.......#.#"),
    list("#.#.#####.###.#"),
    list("#...........#.#"),
    list("###.#.#####.#.#"),
    list("#...#.....#.#.#"),
    list("#.#.#.###.#.#.#"),
    list("#.....#...#.#.#"),
    list("#.###.#.#.#.#.#"),
    list("#S..#.....#...#"),
    list("###############")
]

# Bestimme Start- und Endpunkte
start = (13, 1)  # S befindet sich bei (12, 0)
end = (1, 13)    # E befindet sich bei (1, 12)

# Alle möglichen Wege finden
all_paths = find_all_paths(maze, start, end)

# Ausgabe der gefundenen Wege
print(f"Gefundene Wege von S nach E: {len(all_paths)}")
# for path in all_paths:
#     print(path)
```

```python
from copy import deepcopy

file_name = "test.txt"
file = open(file_name, "r")
data = file.read().strip().split("\n")

G = [[c for c in line] for line in data]
R =len(G)
C = len(G[0])

for r in range(R):
    for c in range(C):
        if G[r][c] == "S":
            start = (r,c)
        if G[r][c] == "E":
            end = (r,c)


def find_all_paths(maze, start, end):
    # Definiere die möglichen Bewegungsrichtungen (nach oben, unten, links, rechts)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    def is_valid_move(x, y):
        # Prüfe, ob der nächste Schritt innerhalb des Labyrinths und nicht auf einer Wand ist
        return 0 <= x < len(maze) and 0 <= y < len(maze[0]) and maze[x][y] != '#'

    # Stack für die iterative DFS
    stack = [(start, [start])]  # (Aktuelle Position, Aktueller Pfad)
    stack = deque(stack)
    paths = []
    visited = set()  # Hilfsstruktur für besuchte Felder, die temporär genutzt wird

    while stack:
        (x, y), path = stack.pop()  # Pop die letzte Position und den Pfad

        # Wenn das Ziel erreicht ist, speichern wir den aktuellen Pfad
        if (x, y) == end:
            paths.append(path)
            continue

        # Markiere den aktuellen Punkt als besucht (temporär)
        visited.add((x, y))

        # Versuche, alle benachbarten Felder zu besuchen
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if is_valid_move(nx, ny) and (nx, ny) not in visited:
                stack.append(((nx, ny), path + [(nx, ny)]))  # Füge neuen Punkt zum Stack hinzu

        # Wir machen keinen Backtracking-Mechanismus direkt hier,
        # da die besuchten Felder in `visited` gespeichert werden.
        # In der nächsten Iteration wird es automatisch überprüft.

    return paths


# Alle möglichen Wege finden
all_paths = find_all_paths(G, start, end)

# Ausgabe der gefundenen Wege
print(f"Gefundene Wege von S nach E: {len(all_paths)}")
for path in all_paths:
    H = deepcopy(G)
    print(path)
    for r in range(R):
        for c in range(C):
            if (r,c) in path:
                H[r][c] = "@"

    for r in range(R):
        print("".join(H[r]))
    print( )
```
